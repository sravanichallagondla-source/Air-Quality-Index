### Fill missing values of numerical data

```python
a = a.loc[a['categori'] != 'TIDAK ADA DATA'].copy()
```

### Handle Missing Values numerical values

```python
pollutants = ['pm25', 'pm10', 'so2', 'co', 'o3', 'no2']

for col in pollutants:
    a.loc[:, col] = a[col].fillna(a[col].median())
```

### Handling Missing values for categorical values

```python
a.loc[:, 'categori'] = a['categori'].fillna(a['categori'].mode()[0])
```

### Fix categorical inconsistencies

```python
a.loc[:, 'categori'] = a['categori'].str.upper().str.strip()
a.loc[:, 'critical'] = a['critical'].str.upper().str.strip()
```

### Convert data types

```python
a['tanggal'] = pd.to_datetime(a['tanggal'])
```

# UNIVARIATE ANALYSIS

```python
### AQI Categorical Distribution
sns.countplot(data=a, x='categori')
plt.title("Distribution of AQI Category in ispu_dki5")
plt.show()
```

## Insight:
- The majority of air quality records fall under Moderate and Unhealthy categories.
- This indicates that Jakarta frequently experiences insafe air quality levels.
- Good air quality days are comparatively fewer. 

```python
### PM2.5 Distribution
sns.histplot(a['pm25'],bins=30,kde=True)
plt.title("PM2.5 Distribution")
plt.show()
```

### Insight:
- PM2.5 values are right-skewed,indicating frequent high pollution events.
- Most PM2.5 readings are concentrated in the medium to high range.
- Fine particulate matter ia a major concern for air quality in Jakarta.

```python
### Dominant Pollutant Count
sns.countplot(x='critical',data=a)
plt.title('PM2.5 Distribution')
plt.show()
```

### Insight:
- PM2.5 and PM10 are the most dominant pollutants in the dataset.
- These pollutants contribute the most to poor air quality conditions.
- Other pollutants like SO2 and CO occur less frequently as critical pollutants.

# Time Series Analysis

```python
### Yearly Trend Analysis
a['tanggal']=pd.to_datetime(a['tanggal'])
a['year'] = a['tanggal'].dt.year

a.groupby('year')['pm25'].mean().plot()
plt.title("Yearly Average PM2.5 Levels")
plt.xlabel('Year')
plt.ylabel('Average PM2.5')
plt.show()
```

### Insight:
- The yearly average PM2.5 levels show fluctuations across different years.
- There is no consistent long-term decline in pollution levels.
- PM2.5 remains a major air quality concern in Jakarta.

```python
### Monthly PM2.5 Variation
a['month']=a['tanggal'].dt.month
sns.boxplot(x='month',y='pm25',data=a)
plt.title('Monthly PM2.5 Variation in Jakarta')
plt.xlabel('Month')
plt.ylabel('Pm2.5')
plt.show()
```

### Insight:
- Pm2.5 levels vary significantly across different months.
- Certain months show higher pollution levels,indicating seasonal effects.
- Weather conditions influence air quality in Jakarta.

```python
### AQI Categorical VS Dominant Pollutant
sns.countplot(x='critical', hue='categori', data=a)
plt.title("Dominant Pollutant vs AQI Category")
plt.xlabel("Dominant Pollutant")
plt.ylabel("Count")
plt.show()
```

### Insight:
- PM2.5 and PM10 are the dominant pollutants across most AQI categories.
- Higher AQI categories are strongly associated with particulate matter.
- Particulate pollution is the primary driver of poor air quality

# Bivariate Analysis

```python
### Categorical Pollutant Vs Categorical
### Dominant Pollutant Vs AQI Category
sns.countplot(x='critical', hue='categori', data=a)
plt.title("Dominant Pollutant vs AQI Category")
plt.xlabel("Dominant Pollutant")
plt.ylabel("Count")
plt.show()
```

### Insight:
- PM2.5 and PM10 are dominant pollutants across most AQI categories
- Higher AQI categories are strongly associated with particulate matter.
- Other pollutants contribute less to sereve air quality conditions.

```python
### Numerical Vs Numerical
### PM2.5 Vs Relationship
sns.scatterplot(x='pm25', y='pm10', data=a)
plt.title("PM2.5 vs PM10")
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.show()
```

### Insight:
- PM2.5 and Pm10 show a positive relationship.
- Higher PM2.5 values are often associated with higher PM10 values.
- This suggests common pollution sources such as traffic and industry.

```python
### Categorical Vs Numerical
### PM2.5 across AQI Categories
sns.boxplot(x='categori', y='pm25', data=a)
plt.title("PM2.5 Distribution across AQI Categories")
plt.xlabel("AQI Category")
plt.ylabel("PM2.5")
plt.show()
```

### Insight:
- Higher AQI categories show higher median PM2.5 levels.
- The spread of PM2.5 increases as air quality worsens.
- Pm2.5 strongly influence AQI classification.

```python
#### Correlation Heatmap
pollutants = a[['pm25', 'pm10', 'so2', 'co', 'o3', 'no2']]
corr_matrix = pollutants.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Air Pollutants in Jakarta")
plt.show()
```

### Insight:
- “PM2.5 and PM10 show a positive but moderate correlation, suggesting some shared pollution sources.”
- This indicates that both pollutants likely originate from similar sources.
- Other pollutants have weaker correlations with particular matter.

# Multivariate Analysis

```python
### Multiple Pollutants Trend Over Years
### Yearly average of multiple pollutants
a.groupby('year')[['pm25', 'pm10', 'no2']].mean().plot(figsize=(8,5))
plt.title("Yearly Trend of Major Pollutants in Jakarta")
plt.xlabel("Year")
plt.ylabel("Average Concentration")
plt.show()
```

### Insight:
- PM2.5 remains consistently higher than other pollutants across years.
- PM10 and NO2 fluctuate but contribute less compared to PM2.5.
- Particulate matter is the dominant air pollution concern in Jakarta.

```python
### AQI Category Vs Pollutants
sns.boxplot(x='categori', y='pm25', data=a)
plt.title("PM2.5 across AQI Categories")
plt.show()
```

### Insight:
- Higher AQI categories show higher PM2.5 concentrations.
- PM2.5 levels increase as air quality worsens.
- AQI classifivation is strongly influenced by PM2.5

```python
### Pollutants by Station
plt.figure(figsize=(10,5))
sns.barplot(x='stasiun', y='pm25', data=a, estimator='mean')
plt.xticks(rotation=45)
plt.title("Average PM2.5 by Monitoring Station")
plt.show()
```

### Insight:
- PM2.5 levels vary across different monitoring stations.
- Certain stations consistently record higher pollution levels.
- Local factor influence air quality measurements.

# Final Insights:
- PM2.5 consistently shows higher concentration levels compared to other pollutants, indicating it as a dominant contributor to air pollution.
- PM2.5 and PM10 show a weak positive correlation, suggesting that while they may share some common pollution sources such as traffic or industrial activities, their relationship is not very strong.
- Air pollution levels fluctuate over time, but no consistent long-term decline is observed.
